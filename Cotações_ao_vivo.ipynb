```python
import yfinance as yf

# Obter o preço em tempo real da Apple
ticker = yf.Ticker("BTC-USD")
preco_atual = ticker.info['regularMarketPrice']
print(preco_atual)
```

### Importando bibliotecas time, datetime e pandas

```python
import time
import datetime
import pandas as pd
```

####Quantidade de minutos que vai ser me passado a cotação.

```python
n = 1
```

```python
tempo = time.time() + 60*n
```

##Programa informando o preço a cada 5 segundos.

```python
a = 0
df = pd.DataFrame(columns = ['horas', 'Preço'])
while time.time() < tempo:
  preco_atual = ticker.info['regularMarketPrice']
  tempoReal = datetime.datetime.fromtimestamp(time.time())
  tempoReal = tempoReal.strftime('%d/%m/%Y %H:%M:%S')
  d = {'horas': tempoReal, 'Preço': preco_atual}
  linha = pd.DataFrame(data = d, index = [a] )
  a += 1
  df = pd.concat([df, linha], ignore_index= True)
  time.sleep(5)
  print(d)
print('Período de cotação finalizado')
```
